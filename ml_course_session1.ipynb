## 2. Supervised vs. Unsupervised Learning

Machine learning algorithms can be broadly categorized into two main paradigms based on the nature of the learning signal or feedback available to the learning system.

image.png

### 2.1 Supervised Learning (Focus of This Course)

**Definition:** Supervised learning involves training a model on labeled data, where each training example consists of an input (features) and a corresponding output (label or target).

**Mathematical Formulation:**
- Given: Training dataset $D = \{(x_1, y_1), (x_2, y_2), ..., (x_n, y_n)\}$
- Where: $x_i$ represents input features and $y_i$ represents the corresponding label
- Goal: Learn a function $f: X \rightarrow Y$ such that $f(x_i) \approx y_i$

**Types of Supervised Learning:**

1. **Regression:** Predicting continuous numerical values
   - Target variable is continuous (e.g., price, temperature, age)
   - Examples: Linear Regression, Polynomial Regression, SVR
   
2. **Classification:** Predicting discrete categories or classes
   - Target variable is categorical (e.g., spam/not spam, species type)
   - Examples: Logistic Regression, Decision Trees, SVM, K-NN

**Supervised Learning Workflow:**
```

 Labeled     
 Training    
 Data        

       
       
     
  Learning    >  Trained Model
  Algorithm          f(x)       
     
                           
           
   New                  
   Input                
   Data                 
           
                         
       
                
                
         
          Predictions 
         
```

### 2.2 Unsupervised Learning (Not Covered in This Course)

**Definition:** Unsupervised learning involves training a model on unlabeled data, where the system tries to learn patterns, structure, or relationships without explicit guidance.

**Characteristics:**
- No labeled outputs provided
- Model discovers hidden patterns or intrinsic structures
- Goal: Find meaningful structure in data

**Common Tasks:**
1. **Clustering:** Grouping similar data points (e.g., customer segmentation)
2. **Dimensionality Reduction:** Reducing number of features while preserving information
3. **Anomaly Detection:** Identifying unusual patterns
4. **Association Rules:** Discovering relationships between variables

### 2.3 Key Differences

| Aspect | Supervised Learning | Unsupervised Learning |
|--------|--------------------|-----------------------|
| **Data** | Labeled (input + output) | Unlabeled (input only) |
| **Goal** | Predict output for new inputs | Discover patterns/structure |
| **Feedback** | Direct (known correct answers) | Indirect (no correct answers) |
| **Complexity** | Generally less complex | Often more complex |
| **Applications** | Prediction, Classification | Exploration, Clustering |

## 3. Understanding Data in Machine Learning

### 3.1 Data Structure in Machine Learning

Machine learning data is typically organized in a tabular format:

```
        Feature 1  Feature 2  Feature 3  ...  Feature n    Target
Sample 1    x       x        x     ...     x        y
Sample 2    x       x        x     ...     x        y
Sample 3    x       x        x     ...     x        y
   ...      ...       ...        ...     ...     ...       ...
Sample m    x       x        x     ...     x        y
```

### 3.2 Key Terminology

**Features (Independent Variables, Predictors, Input Variables):**
- Measurable properties or characteristics used to make predictions
- Denoted as $X$ or individual features as $x_1, x_2, ..., x_n$
- Each row in the feature matrix is called a **sample** or **observation**
- Each column is a **feature** or **attribute**

**Target (Dependent Variable, Label, Output Variable):**
- The variable we want to predict
- Denoted as $y$ or $Y$
- In regression: continuous values
- In classification: discrete categories

**Matrix Notation:**
- Feature matrix: $X \in \mathbb{R}^{m \times n}$ where $m$ = number of samples, $n$ = number of features
- Target vector: $y \in \mathbb{R}^{m}$ (regression) or $y \in \{c_1, c_2, ..., c_k\}$ (classification)

### 3.3 Types of Data

**1. Numerical Data:**
- **Continuous:** Can take any value within a range (e.g., height, temperature, price)
- **Discrete:** Can take only specific values (e.g., number of rooms, age in years)

**2. Categorical Data:**
- **Nominal:** Categories without inherent order (e.g., color, city, species)
- **Ordinal:** Categories with meaningful order (e.g., education level, satisfaction rating)

**3. Binary Data:**
- Special case of categorical with only two categories (e.g., Yes/No, True/False)

### 3.4 Data Quality Considerations

**Common Data Issues:**
1. **Missing values:** Some entries are not recorded
2. **Outliers:** Extreme values that deviate significantly from other observations
3. **Inconsistent formats:** Different units or representations
4. **Imbalanced classes:** Unequal representation of different categories (classification)
5. **High dimensionality:** Too many features relative to samples

### 7.2 Why Split Data?

**Training Set:**
- Used to fit the model parameters
- Model learns patterns from this data
- Typically 70-80% of total data

**Test Set:**
- Used to evaluate model performance
- Model has never seen this data during training
- Simulates real-world deployment
- Typically 20-30% of total data

**Critical Principle:** **Never train on test data!** This leads to **overfitting** and unrealistic performance estimates.

### 7.3 Implementing Train-Test Split

````python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Prepare data for the California Housing dataset
X_housing = housing.data  # Features (all columns except target)
y_housing = housing.target  # Target (what we want to predict)

print("Original dataset shape:")
print(f"Features (X): {X_housing.shape}")
print(f"Target (y): {y_housing.shape}")

print("\n" + "="*50 + "\n")

# Split the data
# test_size=0.2 means 20% for testing, 80% for training
# random_state ensures reproducibility (same split every time)
X_train, X_test, y_train, y_test = train_test_split(
    X_housing, y_housing,
    test_size=0.2,  # 20% for testing
    random_state=42  # For reproducibility
)

print("After splitting:")
print(f"Training set - Features: {X_train.shape}, Target: {y_train.shape}")
print(f"Test set - Features: {X_test.shape}, Target: {y_test.shape}")

print("\n" + "="*50 + "\n")

# Verify the split percentages
total_samples = len(X_housing)
train_percentage = (len(X_train) / total_samples) * 100
test_percentage = (len(X_test) / total_samples) * 100

print(f"Training set: {train_percentage:.1f}% of data")
print(f"Test set: {test_percentage:.1f}% of data")
````

### 7.4 Introduction to Evaluation Metrics

After training a model, we need to measure its performance. Different types of problems require different metrics.

#### Regression Metrics

**1. Mean Squared Error (MSE):**
$$MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$$

- Average of squared differences between predictions and actual values
- Units: squared units of target variable
- Lower is better (0 is perfect)

image.png
**2. Root Mean Squared Error (RMSE):**
$$RMSE = \sqrt{MSE} = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}$$

- Square root of MSE
- Units: same as target variable (more interpretable)
- Lower is better

**3. R Score (Coefficient of Determination):**
$$R^2 = 1 - \frac{\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}{\sum_{i=1}^{n}(y_i - \bar{y})^2}$$

- Proportion of variance in target explained by features
- Range: (-, 1], where 1 is perfect prediction
- 0 means model is as good as predicting the mean
- Negative values indicate worse than baseline

image.png

#### Classification Metrics (Brief Introduction)

**Accuracy:**
$$Accuracy = \frac{\text{Number of Correct Predictions}}{\text{Total Number of Predictions}}$$

- Simple proportion of correct predictions
- Range: [0, 1], where 1 is perfect
- Can be misleading with imbalanced classes

## 8. Your First ML Model: Simple Linear Regression

### 8.1 What is Linear Regression?

Linear Regression models the relationship between features and a continuous target variable using a linear function.

**Mathematical Formula (Simple Linear Regression with one feature):**
$$\hat{y} = \beta_0 + \beta_1 x$$

Where:
- $\hat{y}$ = predicted value
- $x$ = input feature
- $\beta_0$ = intercept (y-intercept)
- $\beta_1$ = slope (coefficient)

image.png

**Multiple Linear Regression (general form):**
$$\hat{y} = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n$$

**Objective:** Find the values of $\beta_0, \beta_1, ..., \beta_n$ that minimize the prediction error.

**Learning Process:** The algorithm minimizes the Mean Squared Error:
$$\min_{\beta_0, \beta_1, ..., \beta_n} \sum_{i=1}^{m}(y_i - \hat{y}_i)^2$$

### 8.2 Simple Linear Regression: One Feature

Let's start with the simplest case: predicting house price using only median income.

````python
# Import LinearRegression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# For simple linear regression, we'll use only one feature: Median Income
# Reshape is needed because sklearn expects 2D arrays
X_simple = X_train[:, 0].reshape(-1, 1)  # First feature (MedInc), reshaped to 2D
X_simple_test = X_test[:, 0].reshape(-1, 1)

print("Training with single feature (Median Income)")
print(f"Training set shape: {X_simple.shape}")
print(f"Test set shape: {X_simple_test.shape}")

print("\n" + "="*50 + "\n")

# Step 1: Create the model
model_simple = LinearRegression()
print(" Model created")

# Step 2: Train the model (fit to training data)
model_simple.fit(X_simple, y_train)
print(" Model trained")

# Step 3: Examine learned parameters
print("\n" + "="*50 + "\n")
print("Learned Parameters:")
print(f"Intercept (): {model_simple.intercept_:.4f}")
print(f"Slope (): {model_simple.coef_[0]:.4f}")
print(f"\nEquation: Price = {model_simple.intercept_:.4f} + {model_simple.coef_[0]:.4f}  MedInc")
````

````python
# Step 4: Make predictions
y_pred_train = model_simple.predict(X_simple)
y_pred_test = model_simple.predict(X_simple_test)

print("First 5 predictions vs actual values (Test Set):")
comparison = pd.DataFrame({
    'MedInc': X_simple_test[:5, 0],
    'Actual Price': y_test[:5],
    'Predicted Price': y_pred_test[:5],
    'Error': y_test[:5] - y_pred_test[:5]
})
print(comparison)

print("\n" + "="*50 + "\n")

# Step 5: Evaluate the model
train_rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
train_r2 = r2_score(y_train, y_pred_train)
test_r2 = r2_score(y_test, y_pred_test)

print("Model Performance:")
print(f"\nTraining Set:")
print(f"  RMSE: {train_rmse:.4f} (in $100k)")
print(f"  R Score: {train_r2:.4f}")

print(f"\nTest Set:")
print(f"  RMSE: {test_rmse:.4f} (in $100k)")
print(f"  R Score: {test_r2:.4f}")

print("\n" + "="*50 + "\n")
print(" Interpretation:")
print(f"   - The model explains {test_r2*100:.1f}% of variance in house prices")
print(f"   - Average prediction error: ${test_rmse*100:.1f}k")
print(f"   - Test RMSE  Train RMSE suggests good generalization (no overfitting)")
````

````python
# Visualize the model
plt.figure(figsize=(12, 5))

# Plot 1: Training data with regression line
plt.subplot(1, 2, 1)
plt.scatter(X_simple, y_train, alpha=0.3, s=10, label='Actual')
plt.plot(X_simple, y_pred_train, color='red', linewidth=2, label='Predicted (Regression Line)')
plt.xlabel('Median Income')
plt.ylabel('Price ($100k)')
plt.title('Training Set: Linear Regression Fit')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2: Test data with regression line
plt.subplot(1, 2, 2)
plt.scatter(X_simple_test, y_test, alpha=0.3, s=10, label='Actual', color='green')
plt.plot(X_simple_test, y_pred_test, color='red', linewidth=2, label='Predicted (Regression Line)')
plt.xlabel('Median Income')
plt.ylabel('Price ($100k)')
plt.title('Test Set: Model Performance')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n Visualization shows the linear relationship learned by the model.")
print("   Points far from the line represent larger prediction errors.")
````

## 9. Introduction to Classification

Now let's briefly explore a classification problem to understand the parallel workflow.

### 9.1 Classification vs Regression

| Aspect | Regression | Classification |
|--------|-----------|----------------|
| **Target Type** | Continuous | Discrete (Categories) |
| **Output** | Numerical value | Class label |
| **Example** | Predicting house price | Predicting flower species |
| **Evaluation** | RMSE, R, MSE | Accuracy, Precision, Recall |

### 9.2 Logistic Regression for Classification

Despite its name, Logistic Regression is a **classification** algorithm. It predicts the probability of a sample belonging to a particular class.

````python
# Prepare Iris data
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

X_iris = iris.data
y_iris = iris.target

# Split the data
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(
    X_iris, y_iris, test_size=0.2, random_state=42
)

print("Iris Dataset Split:")
print(f"Training set: {X_train_iris.shape[0]} samples")
print(f"Test set: {X_test_iris.shape[0]} samples")

print("\n" + "="*50 + "\n")

# Create and train the model
model_classification = LogisticRegression(max_iter=200, random_state=42)
model_classification.fit(X_train_iris, y_train_iris)
print(" Classification model trained")

# Make predictions
y_pred_iris = model_classification.predict(X_test_iris)

# Evaluate
train_accuracy = model_classification.score(X_train_iris, y_train_iris)
test_accuracy = accuracy_score(y_test_iris, y_pred_iris)

print("\n" + "="*50 + "\n")
print("Model Performance:")
print(f"Training Accuracy: {train_accuracy:.4f} ({train_accuracy*100:.2f}%)")
print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
````

````python
# Show some predictions
print("\nFirst 10 predictions vs actual:")
prediction_comparison = pd.DataFrame({
    'Actual': [iris.target_names[y] for y in y_test_iris[:10]],
    'Predicted': [iris.target_names[y] for y in y_pred_iris[:10]],
    'Correct': y_test_iris[:10] == y_pred_iris[:10]
})
print(prediction_comparison.to_string(index=False))

print("\n" + "="*50 + "\n")

# Detailed classification report
print("Detailed Classification Report:")
print(classification_report(y_test_iris, y_pred_iris, target_names=iris.target_names))

print("\n Note: We'll explore these metrics (precision, recall, f1-score) in detail in Session 3.")
````

````python
# Visualize classification results
plt.figure(figsize=(12, 5))

# Plot with actual labels
plt.subplot(1, 2, 1)
for i, color in enumerate(['red', 'blue', 'green']):
    mask = y_test_iris == i
    plt.scatter(X_test_iris[mask, 2], X_test_iris[mask, 3],
                c=color, label=iris.target_names[i], s=100, alpha=0.6, edgecolors='black')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Actual Species')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot with predicted labels
plt.subplot(1, 2, 2)
for i, color in enumerate(['red', 'blue', 'green']):
    mask = y_pred_iris == i
    plt.scatter(X_test_iris[mask, 2], X_test_iris[mask, 3],
                c=color, label=iris.target_names[i], s=100, alpha=0.6, edgecolors='black')

# Mark incorrect predictions
incorrect = y_test_iris != y_pred_iris
plt.scatter(X_test_iris[incorrect, 2], X_test_iris[incorrect, 3],
            s=200, facecolors='none', edgecolors='red', linewidths=3, label='Misclassified')

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Predicted Species')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n Red circles indicate misclassified samples.")
````
